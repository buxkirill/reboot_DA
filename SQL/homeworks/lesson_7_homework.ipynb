```python
import sqlite3
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
conn = sqlite3.connect('task1_7.db')  
c = conn.cursor()
```

### --task1  (lesson7)
-- sqlite3: Сделать тестовый проект с БД (sqlite3, project name: task1_7). В таблицу table1 записать 1000 строк с случайными значениями (3 колонки, тип int) от 0 до 1000.
-- Далее построить гистаграмму распределения этих трех колонко


```python
request = """create table table1 (col1 int, col2 int, col3 int)"""
c.execute(request)
values = [(str(random.randint(0, 1000)), str(random.randint(0, 1000)), str(random.randint(0, 1000))) for _ in range(1000)]
c.executemany("insert into table1 (col1, col2, col3) values (?, ?, ?);", values)
c.execute("select * from table1").fetchall()
```

```python
df = pd.read_sql_query('select * from table1', conn)
for col in df.columns:
  plt.figure(figsize=(10,7))
  sns.distplot(df[col], color='g')
  plt.title(f'Гистограмма распределения для {col}')
  plt.show()
```
